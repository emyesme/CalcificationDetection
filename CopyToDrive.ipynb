```python
import os
import shutil
import time

patchesDisco ="D:\\CalcificationDetection\\Patches"

patchesDrive ="J:\\.shortcut-targets-by-id\\1X3JbrfZbPs0AMWFj_sLDSIkzqHVjU8yp\\Results\\Patches"


contenidos=os.listdir(patchesDisco)
contenidos = [i for i in contenidos if i != "Results_csv"]
for imageKey in contenidos:
    windows = os.listdir(os.path.join(patchesDisco, imageKey))
    flag = False
    for window in windows:
        src = os.path.join(patchesDisco, imageKey, window) # origen
        dst = os.path.join(patchesDrive, imageKey, window) # destino
        if os.path.exists(dst):
            print("Already exists...")
            continue
        else:
            try:
                print("copying {}".format(src))
                shutil.copytree(src, dst)
                flag = True
            except Exception as e:
                print("Error copying {}. Error {}".format(src, e))

#     if flag:
#         time.sleep(60) 
```
