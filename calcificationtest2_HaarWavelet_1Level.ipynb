```python
#Wavelet Transform


#### haar wavelet 
#Appl. Sci. 2019, 9, 5388; doi:10.3390/app9245388
#Apply GLCM on LH, HL, HH

#import python wavelet package

import pywt

img2=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #convert image to grayscale format

  #coeffs2 = pywt.dwt2(original, 'bior1.3')
coeffs2 = pywt.dwt2(img2, 'haar') # apply haar wavelet decomposition
(LL, (LH, HL, HH)) = coeffs2 # separate the approximation, horizontal detail (LH), vertial detail (HL), and diagonal detail (HH)

#### haar wavelet


titles = ['Approximation', ' Horizontal detail','Vertical detail', 'Diagonal detail']
fig = plt.figure(figsize=(12, 13))
for i, a in enumerate([LL, LH, HL, HH]):
  ax = fig.add_subplot(2, 2, i + 1)
  #ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
  ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
  ax.set_title(titles[i], fontsize=10)
  ax.set_xticks([])
  ax.set_yticks([])
fig.tight_layout()
plt.show()
    
```

```python
## Wavelet decomposition (2 level)
n = 2
w = 'haar'
coeffs = pywt.wavedec2(img2,wavelet=w,level=n)

# normalize each coefficient array
coeffs[0] /= np.abs(coeffs[0]).max()
for detail_level in range(n):
    coeffs[detail_level + 1] = [d/np.abs(d).max() for d in coeffs[detail_level + 1]]

arr, coeff_slices = pywt.coeffs_to_array(coeffs)
plt.figure(figsize=(20, 20))
#plt.imshow(arr,cmap=plt.cm.gray)
plt.imshow(arr,cmap='gray',vmin=-0.25,vmax=0.75)

plt.show()
```

```python
#Haar wavelet compression
#https://www.youtube.com/watch?v=eJLF9HeZA8I

import numpy as np

## Wavelet Compression
n = 4
w = 'haar'
coeffs = pywt.wavedec2(img2,wavelet=w,level=n)

coeff_arr, coeff_slices = pywt.coeffs_to_array(coeffs)

Csort = np.sort(np.abs(coeff_arr.reshape(-1)))

for keep in (0.01, 0.05, 0.01, 0.005):
   # thresh = Csort[int(np.floor((1-keep)*len(Csort)))]
    
    thresh = Csort[int(np.floor((keep)*len(Csort)))]

    ind = np.abs(coeff_arr) > thresh
    Cfilt = coeff_arr * ind # Threshold small indices
    
    coeffs_filt = pywt.array_to_coeffs(Cfilt,coeff_slices,output_format='wavedec2')
    
    # Plot reconstruction
    Arecon = pywt.waverec2(coeffs_filt,wavelet=w)
    plt.figure(figsize=(12, 13))
    plt.imshow(Arecon.astype('uint8'),cmap='gray')
    plt.axis('off')
    plt.title('keep = ' + str(keep))

#http://databookuw.com/databook.pdf
```
